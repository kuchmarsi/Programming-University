```python
def shift_abc(abc, k):
    return abc[k:] + abc[:k]

def caesar(text, k):
    abc = "абвгдежзийклмнопрстуфхцъчшыьщэюя"
    abc_shifted = shift_abc(abc, k)
    table = str.maketrans({x:y for (x,y) in zip(abc, abc_shifted)})
    return text.translate(table)

def viginer(text, password):
    abc = "абвгдежзийклмнопрстуфхцъчшыьщэюя"
    n = len(password)
    cipher_text = list(text)
    for i in range(n):
        cipher_text[i::n] = caesar(text[i::n], password[i])
    cipher_text = "".join(cipher_text)
    return cipher_text
```

```python
password = [11, 4, 15, 9, 7, 23, 0]
cipher_text = viginer(clean_text1, password)
cipher_text
```

```python
def compute_correlation(cipher_text):
    counts = []
    for k in range(256): #len(cipher_text)):
        count = 0
        for a, b in zip(cipher_text[:-k], cipher_text[k:]):
            if a == b:
                count += 1
        counts.append(count)
    return counts

counts = compute_correlation(cipher_text)
```
